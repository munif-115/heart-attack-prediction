```python
X=X.drop(['creatinine_phosphokinase','ejection_fraction','platelets','serum_sodium','time'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
regressor=RandomForestRegressor()
```

```python
import numpy as np
n_estimators=[int(x) for x in np.linspace(start=100,stop=1300,num=13)]
n_estimators
```

```python
max_depth=[int(x) for x in np.linspace(start=5,stop=35,num=7)]
max_depth
```

```python
min_samples_split=[2,3,5,10,13,15]
```

```python
max_features=['auto','sqrt']
```

```python
min_samples_leaf=[1,2,5,7,9]
```

```python
random_grid={"n_estimators":n_estimators,
            "max_depth":max_depth,
            "min_samples_split":min_samples_split,
            "max_features":max_features,
            "min_samples_leaf":min_samples_leaf}
print(random_grid)
```

```python
from sklearn.model_selection import RandomizedSearchCV
rf=RandomForestRegressor()
rf_random=RandomizedSearchCV(estimator=rf,param_distributions=random_grid,n_iter=10,cv=5,verbose=2,random_state=32, n_jobs=1)
```

```python
rf_random.fit(X_train,y_train)
```

```python
rf_random.best_params_
```

```python
rf_random.best_score_
```

```python
X.head(2)
```

```python
t=rf_random.predict([[75,0,0,1,1.9,1,0]])
t
```

```python
prediction=rf_random.predict(X_test)
```

```python
sns.distplot(prediction-y_test)
```

```python
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, prediction))
print('MSE:', metrics.mean_squared_error(y_test, prediction))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, prediction)))
```

```python
import pickle
file = open('heartattack.pkl', 'wb')

# dump information to that file
pickle.dump(rf_random, file)
```
